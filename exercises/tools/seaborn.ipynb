```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset("titanic")
df.head()
```

## Draw a Histogram
Draw a histogram to visualize the distribution of the age of passengers. Decorate the plot properly.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages on Titanic', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Draw a Bar Plot
Draw a bar plot with error bars to visualize the mean (bar height) and standard deviation (error bar) of the age of passengers for different genders. Decorate the plot properly.

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='sex', y='age', errorbar=('sd'))
plt.title('Mean Age of Passengers by Gender with Standard Deviation', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Mean Age', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Draw a Count Plot
Draw a count plot to visualize the number of passengers with different ticket classes. Decorate the plot properly.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='class', hue='class', palette='viridis', legend=False)
plt.title('Number of Passengers by Ticket Class', fontsize=16)
plt.xlabel('Ticket Class', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Draw a Joint Plot
Draw a joint plot to visualize the relationship between the age and fare of passengers. Decorate the plot properly.

```python
g = sns.jointplot(data=df, x='age', y='fare', kind='scatter', height=7)
g.fig.suptitle('Relationship between Age and Fare', fontsize=16, y=1.03) # Adjust y to prevent overlap
g.set_axis_labels('Age', 'Fare', fontsize=12)
plt.show()
```

## Draw a Box Plot
Draw a box plot to visualize the relationship between passengers' fares and whether they survived or not. Decorate the plot properly.

```python
import seaborn as sns
import matplotlib.pyplot as plt # Also add matplotlib import for general plotting best practice
df = sns.load_dataset("titanic") # Re-load df to ensure it's defined
g = sns.jointplot(data=df, x='age', y='fare', kind='scatter', height=7)
g.fig.suptitle('Relationship between Age and Fare', fontsize=16, y=1.03) # Adjust y to prevent overlap
g.set_axis_labels('Age', 'Fare', fontsize=12)
plt.show()
```

## Draw a Heatmap
Draw a heatmap to visualize the correlation of attributes. Decorate the plot properly.

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Titanic Attributes', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

## Draw a Facet Grid
Draw a facet grid to visualize the distribution of the `alive` attribute with respect to the `who` attribute (on columns) and the `alone` attribute (on rows). Decorate the plot properly.

```python
g = sns.FacetGrid(df, row='alone', col='who', height=4, aspect=1.2)
g.map_dataframe(sns.countplot, x='alive', hue='alive', palette='viridis', order=['yes', 'no'], legend=False)
g.set_axis_labels('Alive', 'Count')
g.set_titles(col_template='Who: {col_name}', row_template='Alone: {row_name}')
g.add_legend(title='Alive Status')
plt.suptitle('Distribution of Alive Status by Who and Alone', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to prevent suptitle overlap
plt.show()
```

## Draw a Pair Plot
Draw a pair plot to visualize the relationship between each pair of attributes. Decorate the plot properly.

```python
g = sns.pairplot(df.select_dtypes(include=['number']))
g.fig.suptitle('Pair Plot of Titanic Numerical Attributes', y=1.02, fontsize=16)
plt.show()
```
