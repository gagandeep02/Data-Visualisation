# Plotly
You should visualize the `gapminder` dataset.

```python
import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = px.data.gapminder()
df.head()
```

## Draw a Histogram
Draw a histogram to visualize the distribution of populations in 2007. Decorate the plot properly.

```python
df_2007 = df[df['year'] == 2007]
fig = px.histogram(df_2007, x='pop',
                   title='Distribution of Population in 2007',
                   labels={'pop': 'Population'})
fig.update_layout(bargap=0.1) # Add some gap between bars
fig.show()
```

## Draw a Line Plot
Draw a line plot to compare the average life expectancy of continents over time. Decorate the plot properly.

```python
avg_lifeExp_continent = df.groupby(['year', 'continent'])['lifeExp'].mean().reset_index()
fig = px.line(avg_lifeExp_continent, x='year', y='lifeExp', color='continent',
              title='Average Life Expectancy of Continents Over Time',
              labels={'lifeExp': 'Average Life Expectancy', 'year': 'Year'})
fig.show()
```

## Draw a Scatter Plot
Draw a scatter plot to compare the GDP and life expectancy of countries in 2007. Decorate the plot properly.

```python
df_2007 = df[df['year'] == 2007]
fig = px.scatter(df_2007, x='gdpPercap', y='lifeExp',
                 hover_name='country', size='pop', color='continent',
                 log_x=True, size_max=60,
                 title='GDP per Capita vs. Life Expectancy in 2007',
                 labels={'gdpPercap': 'GDP per Capita', 'lifeExp': 'Life Expectancy'})
fig.show()
```

## Draw a Bar Plot
Draw a bar plot to compare GDP of countries in 2007. Decorate the plot properly.

```python
df_2007 = df[df['year'] == 2007]
fig = px.bar(df_2007.sort_values('gdpPercap', ascending=False).head(20), x='country', y='gdpPercap',
                 title='Top 20 Countries by GDP per Capita in 2007',
                 labels={'gdpPercap': 'GDP per Capita', 'country': 'Country'})
fig.show()
```

## Draw a Map Plot
Draw a map plot to visualize the population, life expectancy, and GDP of countries in 2007. Decorate the plot properly.

```python
df_2007 = df[df['year'] == 2007]
fig = px.scatter_geo(df_2007,
                    locations='iso_alpha',
                    color='lifeExp', # Color countries based on life expectancy
                    hover_name='country',
                    size='pop', # Size of bubbles based on population
                    hover_data=['gdpPercap'], # Display GDP per Capita on hover
                    animation_frame='year', # Although only 2007, good to keep for consistency or future use
                    projection='natural earth',
                    title='Population, Life Expectancy, and GDP per Capita of Countries in 2007',
                    labels={'lifeExp': 'Life Expectancy', 'pop': 'Population', 'gdpPercap': 'GDP per Capita'})
fig.update_layout(height=600)
fig.show()
```
